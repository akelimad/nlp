```python
import nltk
import pandas as pd
import string
import wikipediaapi
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
```

```python
wiki_wiki = wikipediaapi.Wikipedia('en')
```

```python
mathematicsDoc = wiki_wiki.page('Mathematics').summary
scienceDoc = wiki_wiki.page('Science').summary
physicsDoc = wiki_wiki.page('Physics').summary
```

```python
vectorizer = TfidfVectorizer()
```

```python
vectors = vectorizer.fit_transform([mathematicsDoc, scienceDoc, physicsDoc])
```

```python
feature_names = vectorizer.get_feature_names()
```

```python
dense = vectors.todense()
dense
```

```python
denselist = dense.tolist()
denselist
```

```python
pd.options.display.max_columns = 500
dataframe = pd.DataFrame(denselist, columns = feature_names)
dataframe
```
